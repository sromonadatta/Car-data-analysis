## Data Cleaning

```python
# Check if there is any null values
car.isnull()
```

```python
# Sum of all null values,if present fil it with the mean value of the column
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(),inplace=True)
car
```

```python
car['Make'].value_counts()
```
